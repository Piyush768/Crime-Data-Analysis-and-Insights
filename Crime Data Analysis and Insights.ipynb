This column, titled "Agency Jurisdiction," indicates the place or region where each crime was reported. # ***5.SERIES***


we have taken five different columns for the series

1. **Agency Jurisdiction**


A list of specific US states is represented by "Agency Jurisdiction" where various criminal incidents have been reported. This column gives important details about the locations where these crimes took place. It acts as a crucial point of reference for finding out how reported crimes are distributed among the various states, allowing police and officials to concentrate their efforts to make city more safe. The analysis of this column enables the development of strategies to improve public safety and community well being

2.**Crime Per Capita**

The "Crime Per Capita" column lists various crimes in relation to the total US population. By measuring the frequency of crimes per person in the population, this metric provides a per capita perspective. It offers a more consistent and fair evaluation of crime rates in various areas, allowing for useful comparisons. By using "Crime Per Capita," we can pinpoint areas with high or low crime rates in relation to population size, allowing officials and police to better allocate their resources.

3.**Homicides Percapita**

This homicides per capita shows the number of homicides in the USA as a percentage of the population is shown in the "Homicides Per Capita" column. By measuring homicides per person in the population, this metric offers a per capita perspective. It provides a fair and consistent evaluation of homicide rates in various areas, enabling meaningful comparisons. We can identify areas with higher or lower homicide rates in relation to their population size by using "Homicides Per Capita," which helps with resource allocation and targeted interventions to increase public safety.


4. **Rapes Percapita**

The "Rapes Per Capita" column measures the annual number of rape cases in various US states in relation to the total population of the nation. This per capita measurement enables a fair evaluation of rape rates that takes population size variations into account. It makes it possible to compare states meaningfully and to identify regions with higher or lower rape rates in relation to their population. "Rapes Per Capita" is a vital tool for policymakers and police to pinpoint areas where sexual assault is a particular problem and implement specialized safety measures.


5. **Violent Crimes**

Data on various types of crimes that take place in various US states are compiled in the "Violent Crimes" column. It includes a variety of criminal activities that are distinguished by a threat or potential for physical harm. This column provides a thorough analysis of violent crime rates in various areas, assisting in the evaluation of public safety and the efficiency of law enforcement. Policymakers and law enforcement organizations can identify regions with higher or lower violent crime rates by analyzing "Violent Crimes," enabling targeted interventions and resource allocation to address safety issues.

# ***6.PLOT***

```python
#Plotting graph of Violet crimes from 1975-2023
import matplotlib.pyplot as plt

years = df['report_year']
violent_crimes = df['violent_crimes']

plt.figure(figsize=(12, 6))
plt.plot(years, violent_crimes, marker='o', linestyle='-', color='blue', markersize=5, label='Violent Crimes')

plt.xlabel('Year')
plt.ylabel('Number of Violent Crimes')
plt.title('Trends in Violent Crimes Over the Years')
plt.grid(True)
plt.legend()

plt.show()
```

1. The violent crimes have been the highest in the category from 1975 to 2015
2. From 1975 to 1980 the violent crimes have been more or less above near 12500.
3. There was a rise in in 1980 to 1981 crossing 15000 mark and there was dip in crime from 1982 to 1985. There was a gradual rise from 1986 to 1990 with peak values being reached at 1990.
4. There was a considerable decrease from 1990 to 2005 and thereby remaining consistent from 2005 till 2009 and then gradually rising from 2010 till 2015 achieving 1980 levels of 5000.
